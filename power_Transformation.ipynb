```python
import pandas as pd
```

```python
data=pd.read_csv('/content/diabetes.csv')
data
```

```python
data.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
x=data.iloc[:,0:8]
y=data.iloc[:,8]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(x_train,y_train)
```

```python
y_pred=lr.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
import numpy as np
acc=accuracy_score(y_test,y_pred)
acc
```

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(lr, x, y, cv=5)
print("Cross-validation scores:", scores)
print("Mean cross-validation score:", scores.mean())
```

```python
from sklearn.preprocessing import PowerTransformer
pt=PowerTransformer()
x_train_trans=pt.fit_transform(x_train)
x_test_trans=pt.fit_transform(x_test)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
for column in data.columns:
    sns.distplot(data[column], kde=True)
    plt.show()
```

```python
lr.fit(x_train_trans,y_train)
y_pred=lr.predict(x_test_trans)
acc=accuracy_score(y_test,y_pred)
acc
```

```python

x_train_trans_df = pd.DataFrame(x_train_trans, columns=x.columns)

for column in x_train_trans_df.columns:
    plt.figure(figsize=(8, 4)) # Optional: set figure size
    sns.distplot(x_train_trans_df[column], kde=True)
    plt.title(f'Distribution of {column} after Power Transformation') # Add a title
    plt.show()
```

```python
scores_transformed = cross_val_score(lr, pt.fit_transform(x), y, cv=5)
print("Cross-validation scores after transformation:", scores_transformed)
print("Mean cross-validation score after transformation:", scores_transformed.mean())
```
