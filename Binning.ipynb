```python
import pandas as pd
from sklearn.preprocessing import KBinsDiscretizer
```

```python
data=pd.read_csv('/content/Hotel Reservations.csv')
data
```

```python
data.dropna(inplace=True)
data
from sklearn.model_selection import train_test_split
x=data.drop(['booking_status'],axis=1)
y=data['booking_status']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
bin=KBinsDiscretizer(n_bins=10,encode='ordinal',strategy='quantile')
b=bin.fit_transform(x_train[['avg_price_per_room']])
b1=bin.fit_transform(x_test[['avg_price_per_room']])
```

```python
binned_data=pd.DataFrame(b,columns=['avg_price_per_room'])
binned_data1=pd.DataFrame(b1,columns=['avg_price_per_room'])

```

```python
data_new=pd.concat([binned_data,binned_data1])
data_new=data_new.reset_index(drop=True)
data_new
```

```python
data['avg_price_per_room_binned'] = data_new['avg_price_per_room']
```
