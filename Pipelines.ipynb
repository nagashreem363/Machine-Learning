```python
import pandas as pd
```

```python
data=pd.read_csv('/content/titanic.csv')
data
```

```python
data.isnull().sum()
```

```python
data.drop(['PassengerId','Name','Ticket','Cabin'],inplace=True,axis=1)
```

```python
data
```

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder, LabelEncoder, StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.compose import ColumnTransformer
import pandas as pd
from sklearn.model_selection import train_test_split

preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), ['Age', 'Fare']),
        ('cat', OneHotEncoder(handle_unknown='ignore'), ['Embarked', 'Sex']),
    ],

    remainder='passthrough' )
imputer = SimpleImputer(strategy='most_frequent')
data[['Age', 'Fare']] = imputer.fit_transform(data[['Age', 'Fare']])


steps=[('preprocessor', preprocessor), # Apply column transformations
       ('model', LogisticRegression()) # Then apply the model
       ]

pipe=Pipeline(steps)

x = data.drop(['Survived'], axis=1)
y = data['Survived'] # 'Survived' is already numerical

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42) # Added random_state for reproducibility
pipe.fit(x_train, y_train) # Pipeline fit method usually takes both X and y
```

```python
y_pred = pipe.predict(x_test)
y_pred
```

```python
# prompt: accuracy score to be checked

from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```
