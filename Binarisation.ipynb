```python
import pandas as pd
```

```python
data=pd.read_csv('/content/Hotel Reservations.csv')
data
```

```python
data.drop(['no_of_weekend_nights', 'no_of_week_nights','type_of_meal_plan','no_of_previous_bookings_not_canceled','no_of_previous_cancellations',
          'repeated_guest','no_of_special_requests','arrival_date','market_segment_type','arrival_year'],inplace=True,axis=1)
```

```python
data
```

```python
data['no_of_travellers']=data['no_of_adults']+data['no_of_children']
data
```

```python
data.drop(['no_of_adults','no_of_children'],axis=1,inplace=True)
```

```python
data
```

```python
from sklearn.preprocessing import Binarizer
b=Binarizer(threshold=1,copy=False)
b_fitted=b.fit_transform(data[['no_of_travellers']])
```

```python
b
```

```python
dataa=pd.DataFrame(b_fitted,columns=['no_of_travellers'])
dataa
```
