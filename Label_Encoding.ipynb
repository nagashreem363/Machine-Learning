```python
import pandas as pd
```

```python
data=pd.read_csv('/content/placementdata.csv')
data
```

```python
y=data['PlacementStatus']
y
```

```python
from sklearn.model_selection import train_test_split
y_train,y_test=train_test_split(y,test_size=0.2)
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
le.fit(y_train)
```

```python
encoded_train=le.transform(y_train)
encoded_test=le.transform(y_test)
```

```python
encoded_train_df=pd.DataFrame(encoded_train,columns=['PlacementStatus'],index=y_train.index)
encoded_test_df=pd.DataFrame(encoded_test,columns=['PlacementStatus'],index=y_test.index)
```

```python
df=pd.concat([encoded_train_df,encoded_test_df])
df
```

```python
data['placement_status']=df['PlacementStatus']
data.drop(columns=['PlacementStatus'],inplace=True)
data
```
